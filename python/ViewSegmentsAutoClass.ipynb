````python
import matplotlib.pyplot as plt
import ProcessSegments as PSeg
import os
import pyAudioAnalysis
import numpy as np
from pyAudioAnalysis import audioBasicIO as aIO
from pyAudioAnalysis import audioSegmentation as aS
````

````python
segments = PSeg.process_data()
````

````python
fs,num_a,_,_ = PSeg.load_data()
````

````python
all_segs = []
for seg in segments:
    all_segs.extend(seg)
````

````python
fig = plt.figure()
plt.plot(all_segs,label="raf_ying_onset_offset")
plt.plot(num_a,label="original Pink Pigeon")
plt.legend()
plt.show()
````

````python



path = os.path.join(os.getcwd(), "../mp3", "XC155388-Pink_Pigeon_(song,_Mauritius,_Black_River_Gorge,_nov2012,_2).MP3")

[Fs, x] = aIO.read_audio_file(path)
segments = aS.silence_removal(x, Fs, 0.020, 0.020, smooth_window = 1.0, weight = 0.2, plot = True)

````

````python

segs_th=[]
for seg in segments:
    onset = int(seg[0]*fs)
    offset = int(seg[1]*fs)
    wave = num_a[onset:offset]
    segs_th.extend(wave)
````

````python
fig = plt.figure()
plt.plot(segs_th,label="hard_threshold")

plt.legend()
plt.show()
````

````python
fig = plt.figure()
plt.plot(segs_th,color='pink',label="hard_threshold")
plt.plot(all_segs,color='grey',label="raf_ying_onset_offset")
plt.plot(num_a,color='blue',label="original Pink Pigeon")
plt.legend()
plt.show()
````
